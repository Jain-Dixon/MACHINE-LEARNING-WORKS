```python
medals=["gold","sliver","gold","bronze","gold","silver"]
gold_count=sliver_count=bronze_count=0
for medal in medals:
  if medal=="gold":
    gold_count+=1
  elif medal=="sliver":
    sliver_count+=1
  else:
    bronze_count+=1
print(gold_count,sliver_count,bronze_count)
```
