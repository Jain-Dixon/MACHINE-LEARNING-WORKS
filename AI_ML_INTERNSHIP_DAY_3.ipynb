```python
data=pd.read_csv('/content/Social_Network_Ads.csv')
data
```

```python
data.head()
```

```python
data.info()
```

```python
data.shape
```

```python
data.drop('User ID',axis=1,inplace=True)
```

```python
sns.boxplot(data)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['Gender_Encoded']=le.fit_transform(data['Gender'])
```

```python
data.head()
```

```python
x=data[['Gender_Encoded','Age','EstimatedSalary']]
y=data['Purchased']
```

```python
x.describe()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x_train[['Age','EstimatedSalary']]= sc.fit_transform (x_train[['Age','EstimatedSalary']])
```

```python

x_test[['Age','EstimatedSalary']]=sc.transform (x_test[['Age','EstimatedSalary']])
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
model_lr=lr.fit(x_train,y_train)
y_pred=model_lr.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
print("confusion_matrix:\n",confusion_matrix(y_test,y_pred))
print("Accuracy:",accuracy_score(y_test,y_pred))
print("Precision:",precision_score(y_test,y_pred))
print("Recall:",recall_score(y_test,y_pred))
print("F1_score:",f1_score(y_test,y_pred))
```

#K-NN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
metric_k=[]
neighbors=np.arange(3,15)
for k in neighbors:
  classifier=KNeighborsClassifier(n_neighbors=k)
  classifier.fit(x_train,y_train)
  y_pred=classifier.predict(x_test)
  acc=accuracy_score(y_test,y_pred)
  metric_k.append(acc)


```

```python
metric_k
```

```python
plt.plot(neighbors,metric_k,'-o')
plt.xlabel('k value')
plt.ylabel('Accuracy')
plt.grid()
```

```python
classifier=KNeighborsClassifier(n_neighbors=k)
classifier.fit(x_train,y_train)
y_pred=classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
print("confusion_matrix:\n",confusion_matrix(y_test,y_pred))
print("Accuracy:",accuracy_score(y_test,y_pred))
print("Precision:",precision_score(y_test,y_pred))
print("Recall:",recall_score(y_test,y_pred))
print("F1_score:",f1_score(y_test,y_pred))
```

#SVM

```python
from sklearn.svm import SVC
sv=SVC(kernel='rbf')
sv.fit(x_train,y_train)
y_pred=sv.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
print("confusion_matrix:\n",confusion_matrix(y_test,y_pred))
print("Accuracy:",accuracy_score(y_test,y_pred))
print("Precision:",precision_score(y_test,y_pred))
print("Recall:",recall_score(y_test,y_pred))
print("F1_score:",f1_score(y_test,y_pred))
```

#DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
dt_clf=DecisionTreeClassifier(random_state=42)
dt_clf.fit(x_train,y_train)
y_pred=dt_clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
print("confusion_matrix:\n",confusion_matrix(y_test,y_pred))
print("Accuracy:",accuracy_score(y_test,y_pred))
print("Precision:",precision_score(y_test,y_pred))
print("Recall:",recall_score(y_test,y_pred))
print("F1_score:",f1_score(y_test,y_pred))
```

#RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf=RandomForestClassifier(random_state=42)
rf_clf.fit(x_train,y_train)
y_pred=rf_clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
print("confusion_matrix:\n",confusion_matrix(y_test,y_pred))
print("Accuracy:",accuracy_score(y_test,y_pred))
print("Precision:",precision_score(y_test,y_pred))
print("Recall:",recall_score(y_test,y_pred))
print("F1_score:",f1_score(y_test,y_pred))
```

```python
rf_clf1=RandomForestClassifier(n_estimators=20,max_depth=20,criterion='entropy',min_samples_split=3)
rf_clf1.fit(x_train,y_train)
y_pred=rf_clf1.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
print("confusion_matrix:\n",confusion_matrix(y_test,y_pred))
print("Accuracy:",accuracy_score(y_test,y_pred))
print("Precision:",precision_score(y_test,y_pred))
print("Recall:",recall_score(y_test,y_pred))
print("F1_score:",f1_score(y_test,y_pred))
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import classification_report
rf = RandomForestClassifier(random_state=42)
param_grid = {
    'n_estimators':[50,100,200,300],
    'max_depth':[None,5,10,15,20],
    'min_samples_split':[2,5,10],
    'min_samples_leaf':[1,2,5],
    'max_features':['sqrt','log2'],
}
```

```python
_random_search = RandomizedSearchCV(estimator=rf, param_distributions=param_grid, n_iter=5, cv=5, scoring='accuracy',n_jobs=-1, random_state=42)
```

```python
random_search.fit(x_train,y_train)
best_model=random_search.best_estimator_
y_pred=best_model.predict(x_test)
print("Best Parameters:",random_search.best_params_)
print("\nCLassification Report:\n",classification_report(y_test,y_pred))
```
