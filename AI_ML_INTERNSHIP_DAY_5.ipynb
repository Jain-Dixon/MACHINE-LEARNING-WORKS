```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (Input,Dense, Conv2D, MaxPool2D, Flatten)
from tensorflow.keras.optimizers import SGD
```

```python
import tensorflow as tf
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
x_train = x_train/255
x_test = x_test/255
```

```python
x_train[0]
```

```python
y_train[0]
```

```python
y_train=to_categorical(y_train)
y_test=to_categorical(y_test)
```

```python
y_train[0]
```

```python
model= Sequential()
model.add(Conv2D(32,(3,3),activation='relu',input_shape=(28,28,1)))
model.add(MaxPool2D((2,2)))
model.add(Flatten())
model.add(Dense(100,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model_mnist = model.fit(x_train,y_train,epochs=10,validation_data=(x_test,y_test))
```

```python
y_pred=model.predict(x_test)
```

#Bag-of-words


```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(ngram_range=(1,2))
traindata=cv.fit_transform(corpus)
x=traindata
y=data['v1']
```

```python
corpus = [

          "data science is an overlap between arts an science",
          'generally arts and graduates are right-brained and science graduates are left-brained'

]
```

```python
def vectorizer_text(corpus):
  bow_model = CountVectorizer()
  dense_vec= bow_model.fit_transform(corpus).todense()
  bow_df=pd.DataFrame(dense_vec)
  bow_df.columns=sorted(bow_model.vocabulary_)
  return bow_df
```

```python
df_bow=vectorizer_text(corpus)
df_bow
```

```python
x
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=100)
clf.fit(x,y)
```

```python
from sklearn import metrics
y_pred=clf.predict(x)
metrics.accuracy_score(y,y_pred)
```

```python
def find_spam(input):
  input=cv.transform(input)
  prediction=clf.predict(input)
  if prediction==0:
    print('message is ham')
  if prediction==1:
    print('message is spam')
```

```python
input=["gladwin lottery adichu."]
find_spam(input)
```

#TF_IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
def vectorizer_tfidf(corpus):
  tfidf_model = TfidfVectorizer()
  vector_list= tfidf_model.fit_transform(corpus).todense()
  tfidf_df=pd.DataFrame(vector_list)
  tfidf_df.columns=sorted(tfidf_model.vocabulary_)
  return tfidf_df
```

```python
df1=vectorizer_tfidf(corpus)
df1
```
