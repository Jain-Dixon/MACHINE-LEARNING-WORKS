```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/beer-servings.csv')
data.head()
```

```python
data.tail()
```

```python
data.describe()
```

```python
data.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
data.shape
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
freqgraph= data.select_dtypes(include=['float'])
freqgraph.hist(figsize=(20,15))
plt.show()
```

```python
data['beer_servings'] =data['beer_servings'].fillna(data['beer_servings'].median())
```

```python
for i in [ 'spirit_servings','wine_servings','total_litres_of_pure_alcohol']:data[i]=data[i].fillna(data[i].median())
```

```python
data.isna().sum()
```

```python
sns.boxplot(data)
```

```python
Q1=np.percentile(data['wine_servings'],25,interpolation='midpoint')
Q2=np.percentile(data['wine_servings'],50,interpolation='midpoint')
Q3=np.percentile(data['wine_servings'],75,interpolation='midpoint')
```

```python
IQR = Q3 - Q1
low_lim=Q1 - 1.5*IQR
up_lim=Q3 + 1.5*IQR
```

```python
outlier = []
for x in data['wine_servings']:
  if ((x > up_lim) or (x < low_lim)):
    outlier.append(x)
print(outlier)
```

```python
len(outlier)
```

```python
ind2 = data['wine_servings']>up_lim
ind = data.loc[ind2].index
```

```python
data = data.drop(ind)
```

```python
data.shape
```

```python
data.drop('country',axis=1,inplace=True)
```

```python
data['continent'].nunique()
```

```python
data = pd.get_dummies(data,dtype='int64')
```

```python
data.head()
```
