identifying prime and non prime number

```python
n = int(input("Enter the number: "))

if n < 2:
    print(f"{n} is not a prime number.")
else:
    is_prime = True
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            is_prime = False
            break
    if is_prime:
        print(f"{n} is a prime number.")
    else:
        print(f"{n} is not a prime number.")
```

reversing the sequance

```python
numbers = input("Enter numbers separated by commas: ")
nums = [int(x.strip()) for x in numbers.split(",")]

reversed_list = []
for i in range(len(nums) - 1, -1, -1):
    reversed_list.append(nums[i])

print(",".join(map(str, reversed_list)))
```

```python
numbers = input("Enter three numbers separated by commas: ")
a, b, c = [int(x.strip()) for x in numbers.split(",")]

if a >= b and a >= c:
    print(a)
elif b >= a and b >= c:
    print(b)
else:
    print(c)
```

```python
rows = 5

# Upper half (increasing)
for i in range(1, rows + 1):
    for j in range(i):
        print("*", end="")
    print()

# Lower half (decreasing)
for i in range(rows - 1, 0, -1):
    for j in range(i):
        print("*", end="")
    print()
```
